# cleaning the data

#### Fix Datatype of the column

```python
data['Order Date']=pd.to_datetime(data['Order Date'])
```

```python
data['Ship Date']=pd.to_datetime(data['Ship Date'])
```

```python
data[['Order Date','Ship Date']].dtypes
```

#### Drop duplicatae if Exist

```python
data=data.drop_duplicates()
```

```python
data.shape
```

#### Improve column name replace space with '_'

```python
data.columns=data.columns.str.replace(' ','_')
```

```python
data.columns=data.columns.str.replace('-','_')
```

```python
data.columns.tolist()
```

#### Strip whitespace from text columns

```python
str_cols = data.select_dtypes(include='object').columns
```

```python
data[str_cols]=data[str_cols].apply(lambda x:x.str.strip())
print("Whitespace cleaned from", len(str_cols), "columns")
```

### Feature Engineering

```python
data['Order_month']=data['Order_Date'].dt.month
data['Order_year']=data['Order_Date'].dt.year
data['Order_month_name']=data['Order_Date'].dt.strftime('%b')
data['Order_quarter']=data['Order_Date'].dt.quarter
data['Order_Q_Label']=data['Order_Date'].dt.to_period('Q').astype(str)
```

```python
# Profit_margin 
data['Profit_margin'] = (data['Profit'] / data['Sales']) * 100
```

```python
# Ship_Days - Number of days to complete shipment
data['Ship_Days']=(data['Ship_Date']-data['Order_Date']).dt.days
```

```python
# Revenue bucket — categorise order size
data['Sales_Bucket'] = pd.cut(data['Sales'],
                        bins=[0, 100, 500, 1000, 10000],
                        labels=['Small','Medium','Large','Enterprise'])
```

```python
data['Is_Loss']=np.where(data['Profit']<=0,'Yes','No')
```

# Category Analysis

#### sales by category

```python
data.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title('Category wise Sales')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.savefig('Category-Sales.png', dpi=300, bbox_inches='tight')
plt.show()
```

#### sales by sub category

```python
data.groupby('Sub_Category')['Sales'].sum().head(10).plot(kind='bar')
plt.title('Sub-Category sales')
plt.ylabel('Sales')
plt.xlabel('Sub Category')
plt.xticks(rotation=45)
plt.savefig('sub-category-Sales.png', dpi=300, bbox_inches='tight')
plt.show()
```

#### Profit by Sub-category


```python
subcat_profit = data.groupby('Sub_Category')['Profit'].sum().sort_values(ascending=False)

# color logic: green = profit, red = loss
colors = ['green' if x > 0 else 'red' for x in subcat_profit]

plt.figure(figsize=(12,6))

subcat_profit.plot(kind='bar', color=colors, edgecolor='black')

plt.title('Sub-Category Profit')
plt.xlabel('Sub-Category')
plt.ylabel('Profit')   # ✅ fixed

plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.savefig('Subcatergory_profit.png',dpi=300, bbox_inches='tight')
plt.show()
```

# Tiem base analysis

#### Monthly sales by Year

```python
years = [2014, 2015, 2016, 2017]

for y in years:
    year_data = data[data['Order_year'] == y]
    monthly_sales = year_data.groupby('Order_month_name')['Sales'].sum()
    
    monthly_sales.plot(kind='bar')
    plt.title(f'Monthly Sales - {y}')
    plt.savefig('Monthly_sales_year.png')
    plt.show()
```

# Regional Analysis

####  Sales by Region

```python
data.groupby('Region')['Sales'].sum().plot(kind='pie',autopct='%1.1f%%')
plt.title('Sales Distribution by Region')
plt.ylabel('')   # removes extra label
plt.savefig('Sales_by_region.png')
plt.show()
```

#### Profit by Region

```python
proreg = data.groupby('Region')['Profit'].sum().sort_values(ascending=False)
colors = ['green' if x > 0 else 'red' for x in proreg]

plt.figure(figsize=(8,6))

proreg.plot(kind='bar', color=colors, edgecolor='black')


plt.title('Profit by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('Profit_by_region.png',dpi=300, bbox_inches='tight')
plt.show()
```

#### loass making state

```python

loss=data[data['Profit']<0]
loss_state=loss.groupby('State')['Profit'].sum().sort_values()
loss_state.plot(kind='bar',color='red',edgecolor='black')

plt.ylabel('Profit')
plt.xlabel('State')
plt.title('Loss Making State')
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.tight_layout()
plt.savefig('Loss_making_state.png',dpi=300,bbox_inches='tight')
plt.show()
```
